```python

import cv2
import matplotlib.pyplot as plt
img = cv2.imread("../DATA/lcd_calc.jpg")
cv2.rectangle(img, pt1=(400,400), pt2=(800,800), color=(255,0,0), thickness=cv2.FILLED, lineType=8, shift=0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)

```
