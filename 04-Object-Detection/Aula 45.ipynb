# Canny Algorithm

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../DATA/sammy_face.jpg')
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(img, threshold1=200, threshold2=255)
```

```python
plt.imshow(edges)
```

```python
med_val = np.median(img)
```

```python
med_val
```

```python
lower = int(max(0,0.7*med_val))
upper = int(min(255,1.3*med_val))
```

```python
edges = cv2.Canny(img, threshold1=lower, threshold2=upper+100)
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img, ksize=(5,5))
plt.imshow(blurred_img)
```

```python
edges = cv2.Canny(blurred_img, threshold1=25, threshold2=190)
plt.imshow(edges)
```
